```python
# import hw1_q2
import subprocess
import os
```

```python

filename = 'hw1-q2.py'
learning_rates = ['0.001', '0.01' , '0.1']
hidden = ['200']
dropout = ['0.5']
activation = ['tanh']
layers = ['2', '3']
```

```python
for i in learning_rates:
    print('logistic_regression ' + '-learning_rate=' + i)
    subprocess.run(["python3", filename, 'logistic_regression', '-learning_rate=' + i, '-batch_size=1'])

for i in learning_rates:
    print('mlp ' + '-learning_rate=' + i)
    subprocess.run(["python3", filename, 'mlp', '-learning_rate=' + i])

for i in hidden:
    print('mlp ' +  '-hidden_size=' + i)
    subprocess.run(["python3", filename, 'mlp', '-hidden_size=' + i])

for i in activation:
    print('mlp '+ '-activation=' + i)
    subprocess.run(["python3", filename, 'mlp', '-activation=' + i])

for i in dropout:
    print('mlp ' + '-dropout=' + i)
    subprocess.run(["python3", filename, 'mlp', '-dropout=' + i])
```

```python
for j in layers:
    print(j + ' layers\n')
    for i in learning_rates:
        print('mlp ' + '-learning_rate=' + i)
        subprocess.run(["python3", filename, 'mlp', '-learning_rate=' + i, '-layers=' + j])

    for i in hidden:
        print('mlp ' + '-hidden_size=' + i)
        subprocess.run(["python3", filename, 'mlp', '-hidden_size=' + i, '-layers=' + j])

    for i in activation:
        print('mlp ' + '-activation=' + i)
        subprocess.run(["python3", filename, 'mlp', '-activation=' + i, '-layers=' + j])

    for i in dropout:
        print('mlp ' + '-dropout=' + i)
        subprocess.run(["python3", filename, 'mlp', '-dropout=' + i, '-layers=' + j])
    
```
